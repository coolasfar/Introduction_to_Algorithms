## red-black trees

```python
class Node(object):
    # assume a new node is Red
    RED="red"
    BLACK="black"
    def __init__(self,key,color=RED):
        self.color=color
        self.key=key
        self.left=None
        self.right=None
        self.p=None
    def __str__(self):
        return str(self.key)+" , "+self.color
    
class RB_tree(object):
    def __init__(self):
        self.root=None
        
    def inorder(self,node):
        r=[]
        if node is not None:
            r.extend(self.inorder(node.left))
            r.append((node.key,node.color))
            r.extend(self.inorder(node.right))
        return r
        
    def tree_min(self,node=None):
        if not node:
            node=self.root
        while node.left is not None:
            node=node.left
            
        return node
    def left_rotate(self,x):
        y=x.right
        x.right=y.left
        if y.left:
            y.left.p=x
        y.p=x.p
        if not x.p:
            self.root = y
        elif x==x.p.left:
            x.p.left=y
        else:
            x.p.right=y
        y.left=x
        x.p=y
    
    def right_rotate(self,x):
        y=x.left
        x.left=y.right
        if y.right:
            y.right.p=x
        
        y.p=x.p
        if not x.p:
            self.root=y
        elif x==x.p.left:
            x.p.left=y
        else:
            x.p.right=y
            
        y.right=x
        x.p=y
    
    def RB_insert_fixup(self,z):
        while z!=self.root and z.p.color==Node.RED:
            if z.p == z.p.p.left:
                y=z.p.p.right   # y is z's uncle node
                if y is not None and y.color== Node.RED:
                    #. case 1  if z's father and uncle nodes both red, turn them to black and
                    # turn z's grandpa to red ,move z to its grandpa
                    z.p.color=Node.BLACK
                    y.color = Node.BLACK
                    z.p.p.color=Node.RED
                    z=z.p.p 
                else:
                
                    if z == z.p.right:
                    # case 2 if z is a right node, rotate to left
                        z=z.p
                        self.left_rotate(z)
                    #case 3 if  z is left node
                    z.p.color=Node.BLACK
                    z.p.p.color=Node.RED
                    self.right_rotate(z.p.p)
            else:
                y=z.p.p.left
                if y is not None and y.color == Node.RED:
                    z.p.color= Node.BLACK
                    y.color = Node.BLACK
                    z.p.p.color=Node.RED
                    z=z.p.p
                else:
                    if z==z.p.left:
                        z=z.p
                        self.right_rotate(z)
                    z.p.color=Node.BLACK
                    z.p.p.color=Node.RED
                    self.left_rotate(z.p.p)
        self.root.color=Node.BLACK
        
    def RB_insert(self,z):
        node=Node(z)
        y=None
        x=self.root
        while x != None:
            y=x
            if node.key<x.key:
                x=x.left
            else:
                x=x.right
        node.p=y
        if y == None:
            self.root=node
        elif node.key < y.key:
            y.left = node
        else:
            y.right=node
        node.left=None
        node.right=None
        node.color=Node.RED
        
        self.RB_insert_fixup(node)
                     
    def RB_transplant(self,u,v):
        if u.p == None:
            self.root = v
        elif u == u.p.left:
            u.p.left=v
        else:
            u.p.right = v
        if v is not None:
            v.p = u.p
        
    def RB_delete(self,z):
        y=z
        color=y.color
        if z.left == None:
            x=z.right
            self.RB_transplant(z,z.right)
        elif z.right == None:
            x=z.left
            self.RB_transplant(z,z.left)
        else:
            y=self.tree_min(z.right)
            color=y.color
            x=y.right
            if x !=None and x.p!=None and y.p == z:
                x.p=y
            else:
                self.RB_transplant(y,y.right)
                y.right=z.right
                y.right.p=y
            self.RB_transplant(z,y)
            y.left = z.left
            y.left.p=y
            y.color=z.color
        if x and color == Node.BLACK:
            self.RB_delete_fixup(x)
        
    def RB_delete_fixup(self,x):
        while x != self.root and x.color == Node.BLACK:
            if x == x.p.left:
                w = x.p.right
                if w is not None and w.color == Node.RED:
                    # case 1
                    w.color=Node.BLACK
                    x.p.color= Node.RED
                    self.left_rotate(x.p)
                    w=x.p.right
                if w.left.color == Node.BLACK and w.right.color == Node.BLACK:
                    # case 2
                    w.color = Node.RED
                    x=x.p
                else:
                    if w.right.color == Node.BLACK:
                        # case 3
                        w.left.color = Node.BLACK
                        w.color = Node.RED
                        self.right_rotate(w)
                        w = x.p.right
                    # case 4
                    w.color = x.p.color
                    x.p.color = Node.BLACK
                    w.right.color = Node.BLACK
                    self.left_rotate(x.p)
                    x = self.root
            else:
                w=x.p.left
                if w is not None and w.color == Node.RED:
                    # case 1
                    w.color=Node.BLACK
                    x.p.color= Node.RED
                    self.right_rotate(x.p)
                    w=x.p.left
                if w.right.color == Node.BLACK and w.left.color == Node.BLACK:
                    # case 2
                    w.color = Node.RED
                    x=x.p
                else:
                    if w.left.color == Node.BLACK:
                        # case 3
                        w.right.color = Node.BLACK
                        w.color = Node.RED
                        self.left_rotate(w)
                        w = x.p.left
                    # case 4
                    w.color = x.p.color
                    x.p.color = Node.BLACK
                    w.left.color = Node.BLACK
                    self.right_rotate(x.p)
                    x = self.root
                
        x.color = Node.BLACK
                
            
    
```

关于红黑树的添加有3种情况：
1. 添加结点的父结点和叔结点都是红色。把父叔两个结点都变成黑色，然后祖父结点变为红色，然后指针上移两层（变成祖父）
2. 添加结点的父结点是红色，叔结点是黑色，添加结点是一个右孩子。指针上移一层，左旋转，这时指针变为左孩子。（变为情况3）
3. 添加结点的父结点是红色，叔结点是黑色，添加结点是一个左孩子。父结点变成黑色，祖父变成红色，右旋转。

红黑树的删除有4种情况：
（要删除的结点x有两重黑色）
1. x的兄弟结点w是红色的。改变w和他们父结点的颜色，然后对父结点左旋，现在x的兄弟结点变成了原来w的一个子结点。进入情况2，3，4
2. x的兄弟结点w是黑色的，且w的两个子结点都是黑色。从x和w去掉一重黑色，x为一重黑色，w为红色。x移到父结点
3. x的兄弟结点w是黑色，w左孩子是红色，w右孩子是黑色。w左孩子变成黑色，w变为红色，右旋w。进入情况4
4. x的兄弟结点w是黑色，w右孩子是红色。将w变为父结点的颜色，父结点变为黑色，w右孩子变为黑色，右旋父结点，x指向root

## AVL tree

The heights of the left and right subtrees of x differ at most 1

```python
class Node:
    def __init__(self,key):
        self.key=key
        self.left=None
        self.right=None
    def __str__(self):
        return self.key
class AVLTree:
    def __init__(self):
        self.root=None
        self.height=-1
        self.balance=0
    def insert(self,key):
        n=Node(key)
        
        if self.root==None:
            self.root=n
            self.root.left=AVLTree()
            self.root.right=AVLTree()
        elif key < self.root.key:
            self.root.left.insert(key)
        elif key > self.root.key:
            self.root.right.insert(key)
        
        self.rebalance()
        
    def rebalance(self):
        
        #check if need to rebalance the tree
        self.update_heights(recursive=False)
        self.update_balances(False)
        
        while self.balance < -1 or self.balance > 1:
            if self.balance > 1: #left subtree is larger
                if self.root.left.balance < 0:
                    self.root.left.rotate_left()
                    self.update_heights()
                    self.update_balances()
                self.rotate_right()
                self.update_heights()
                self.update_balances()
            if self.balance < -1:
                if self.root.right.balance > 0:
                    self.root.right.rotate_right()
                    self.update_heights()
                    self.update_balances()
                self.rotate_left()
                self.update_heights()
                self.update_balances()
                
    def update_heights(self,recursive=True):
        if self.root:
            if recursive:
                if self.root.left:
                    self.root.left.update_heights()
                if self.root.right:
                    self.root.right.update_heights()
            self.height= 1+ max(self.root.left.height,self.root.right.height)
        else:
            self.height=-1
        
    def update_balances(self,recursive=True):
        if self.root:
            if recursive:
                if self.root.left:
                    self.root.left.update_balances()
                if self.root.right:
                    self.root.right.update_balances()

            self.balance = self.root.left.height - self.root.right.height
        else:
            self.balance = 0 
            
    def rotate_right(self):
        
        new_root = self.root.left.root
        new_left_sub = new_root.right.root
        old_root = self.root

        self.root = new_root
        old_root.left.root = new_left_sub
        new_root.right.root = old_root

    def rotate_left(self):

        new_root = self.root.right.root
        new_left_sub = new_root.left.root
        old_root = self.root

        self.root = new_root
        old_root.right.root = new_left_sub
        new_root.left.root = old_root
        
    def inorder(self):
        res=[]
        if self.root is not None:
            res.extend(self.root.left.inorder())
            res.append(self.root.key)
            res.extend(self.root.right.inorder())
            
        return res

                    
                
                
            
```
